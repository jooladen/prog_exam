```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 크롬 브라우저 설정
options = Options()
#options.headless = True  # 브라우저 UI 없이 실행 (headless 모드)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# URL 설정
URL = 'https://finance.naver.com/'

# 페이지 로딩
driver.get(URL)

# 페이지의 상한가 종목이 로드될 때까지 대기
try:
    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '#_topItems2 > tr')))
except Exception as e:
    print("페이지 로딩 실패:", e)

# 상한가 종목들 가져오기
units_up = driver.find_elements(By.CSS_SELECTOR, '#_topItems2 > tr')

print('=====금일 상한가 종목=====')
print()

# 상한가 종목 출력
for unit in units_up[:15]:
    try:
        title_up = unit.find_element(By.CSS_SELECTOR, 'th > a').text
        price_up = unit.find_element(By.CSS_SELECTOR, 'td').text
        up = unit.find_element(By.CSS_SELECTOR, 'td:nth-child(3)').text
        percent_up = unit.find_element(By.CSS_SELECTOR, 'td:nth-child(4)').text

        up = up.replace('상한가', '↑')

        print('종목 이름:', title_up)
        print('한 주당 가격:', price_up + '원')
        print('전날 대비 가격 변동:', up)
        print('전날 대비 등락:', percent_up.strip())
        print()
    except Exception as e:
        print("데이터 추출 실패:", e)

# 드라이버 종료
driver.quit()
```
