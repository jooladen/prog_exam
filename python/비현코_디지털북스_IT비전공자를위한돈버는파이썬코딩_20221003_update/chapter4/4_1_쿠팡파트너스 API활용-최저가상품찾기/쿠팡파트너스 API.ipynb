#### 1_1. 쿠팡파트너스 API를 활용해서, 쿠팡의 데이터를 가져온다.

```python
#sollution
#쿠팡파트너스 API를 가져온다.

import hmac
import hashlib
import binascii
import os
import time
import requests
import json
import urllib
import pandas as pd

#검색하고자 하는 키워드를 입력받는다.
keyword1 = input("원하는 키워드를 검색해주세요 : ")
#검색하고자 하는 키워드를 파이썬에서 해석하는 코드로 바꾼다 ( 한글을 파이썬에서 이해하는 언어로 바꾼다.)
keyword = urllib.parse.quote(keyword1)

#기존 API는 POST로 되어있으나, 오늘 실습에서는 키워드 검색에 따른 결과를 가져와야하기 때문에 GET을 변경한다.
REQUEST_METHOD = "GET"
DOMAIN = "https://api-gateway.coupang.com"
#가져올 API를 확인해서 내가 원하는 키워드를 넣는다. 여기서 리미트는 100으로 지정한다.
URL = f"/v2/providers/affiliate_open_api/apis/openapi/products/search?keyword={keyword}&limit=10"

# Replace with your own ACCESS_KEY and SECRET_KEY
# 본인의 SECRET KEY를 집어넣는다.
ACCESS_KEY = "" #수강생의 access_key 를집어넣으세요
SECRET_KEY = "" #수강생의 secret_key 를집어넣으세요

REQUEST = { "coupangUrls": [
    "https://www.coupang.com/np/search?component=&q=good&channel=user", 
    "https://www.coupang.com/np/coupangglobal"
]}


def generateHmac(method, url, secretKey, accessKey):
    path, *query = url.split("?")
    os.environ["TZ"] = "GMT+0"
    datetime = time.strftime('%y%m%d',time.gmtime())+'T'+time.strftime('%H%M%S',time.gmtime())+'Z'
#     dateGMT = strftime()
#     timeGMT = strftime()
    message = datetime + method + path + (query[0] if query else "")

    signature = hmac.new(bytes(secretKey, "utf-8"),
                         message.encode("utf-8"),
                         hashlib.sha256).hexdigest()

    return "CEA algorithm=HmacSHA256, access-key={}, signed-date={}, signature={}".format(accessKey, datetime, signature)


authorization = generateHmac(REQUEST_METHOD, URL, SECRET_KEY, ACCESS_KEY)
url = "{}{}".format(DOMAIN, URL)
resposne = requests.request(method=REQUEST_METHOD, url=url,
                            headers={
                                "Authorization": authorization,
                                "Content-Type": "application/json"
                            },
                            data=json.dumps(REQUEST)
                            ) 

#여기서 발생된 자료를 확인한다.
print(resposne.json())
#확인된 자료에서 우리가 얻고자 하는 자료만 가져온 뒤 , raw_data에 집어넣는다.
raw_data = resposne.json()['data']['productData']
```

#### 5. 각 rank에 해당하는 사진과 엑셀 같은 폴더에 저장한다.

```python
#sollution
#엑셀저장
#엑셀로 해당 키워드폴더안에 test.xlsx 라는 이름으로 저장한다.
df_sort.to_excel(f'{keyword1}/test.xlsx')

#사진저장

# 이미지를 저장하기 위해서 새로운 데이터프레임으로 rank(검색순서) ,  productImage(썸네일 URL) 만 남긴다.
df_sort_image = df_sort[["rank","productImage"]]
# 해당 DataFrame을 list로 바꿔준다. 리스트로 바꿔주는 것은, for문을 통해서 반복작업을 하기위함
df_sort_image_list = list(df_sort_image.iloc)

# url을 이미지로 저장하기 위한 라이브러리를 설치하고
import urllib.request
# 가져온 리스트에서 링크하나하나를 저장한다.
for a in df_sort_image_list:
    url2 = a[1]
    urllib.request.urlretrieve(url2, f'{keyword1}/{a[0]}.jpg')
    print(str(a[0])+"저장완료")
```
