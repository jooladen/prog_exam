```python
import requests
from bs4 import BeautifulSoup

url = "https://news.naver.com/"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

headlines = soup.select("a[href^='/main/read.naver']")
print(headlines)
for idx, line in enumerate(headlines[:10]):
    print(f"{idx+1}. {line.text.strip()}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://news.naver.com/main/ranking/popularDay.naver?officeId=055"  # SBS
headers = {"User-Agent": "Mozilla/5.0"}
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, "html.parser")

titles = soup.select(".rankingnews_box a")

count = 0
for link in titles:
    title_text = link.text.strip()
    if title_text:  # 내용이 있는 링크만 출력
        count += 1
        print(f"{count}. {title_text}")
```
