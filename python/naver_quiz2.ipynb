```python
import requests
from bs4 import BeautifulSoup

url = "https://www.etnews.com/20250415000268"


response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 제목: h2 태그의 ID가 article_title_h2
title_tag = soup.find('h2', id='article_title_h2')
title = title_tag.text.strip() if title_tag else '제목 없음'

# 날짜: <time> 태그 중 "발행일"이 포함된 첫 번째 태그
time_tags = soup.find_all('time')
date = '날짜 없음'
for tag in time_tags:
    if '발행일' in tag.text:
        date = tag.text.replace('발행일 :', '').strip()
        break

# 본문: articleBody 안의 <p> 태그들 텍스트를 모두 이어붙이기
article_body = soup.find('div', id='articleBody')
if article_body:
    paragraphs = article_body.find_all('p')
    content = '\n\n'.join(p.get_text(separator=' ', strip=True) for p in paragraphs)
else:
    content = '내용 없음'

# 출력
print("제목:", title)
print("날짜:", date)
print("\n내용 요약:\n", content[:500], "...")
print("\n내용 요약2:\n", content)
```
