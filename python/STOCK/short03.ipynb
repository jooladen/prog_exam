```python
#  
import yfinance as yf
import matplotlib.pyplot as plt

#   (: 005930.KS /  : AAPL, TSLA )
ticker = "005930.KS"

# yfinance   ( 3)
df = yf.download(ticker, period="3mo")

#  
print(df.head())

#  
plt.figure(figsize=(10, 4))
plt.plot(df["Close"], label=" (Close)")
plt.title("  3 ")
plt.xlabel("")
plt.ylabel(" ()")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
