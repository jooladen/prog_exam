```python
# 패키지 불러오기
import yfinance as yf
import matplotlib.pyplot as plt

# 삼성전자 티커 (한국: 005930.KS / 미국 예: AAPL, TSLA 등)
ticker = "005930.KS"

# yfinance로 데이터 다운로드 (최근 3개월)
df = yf.download(ticker, period="3mo")

# 데이터 확인
print(df.head())

# 종가 시각화
plt.figure(figsize=(10, 4))
plt.plot(df["Close"], label="종가 (Close)")
plt.title("삼성전자 최근 3개월 주가")
plt.xlabel("날짜")
plt.ylabel("가격 (원)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
